# Calculating and plotting frequency distributions

```python
fdist = nltk.FreqDist(w.lower() for w in brown.words())
```

```python
fdist
```

```python
fdist.values()
```

```python
high_freqs = sorted(fdist.values(), reverse=True)[:100]
```

```python
high_freqs
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot
matplotlib.pyplot.plot(high_freqs)
```
